```python
df_ob2=df.iloc[:,0:22]
df_ob2.drop(['health_insurance'], axis=1, inplace=True)
```

```python
## feature engineering / Handling categorical values ------
from sklearn.preprocessing import OneHotEncoder
df_ob1=df.iloc[:,22:31]
df_ob1=df_ob1.drop(['income_poverty','hhs_geo_region','education'],axis=1)
df_ob1['sex']=df_ob1['sex'].map({'Female':0, 'Male':1})
df_ob1['marital_status']=df_ob1['marital_status'].map({'Not Married':0, 'Married':1})
df_ob1['rent_or_own']=df_ob1['rent_or_own'].map({'Rent':0, 'Own':1})

## encoding for race column 
encoder = OneHotEncoder(categories='auto', sparse=False)
data_encoded = encoder.fit_transform(df_ob1[['race']])
categories = encoder.categories_[0]
encoded_df = pd.DataFrame(data_encoded, columns=categories)
df_ob1 = pd.concat([df_ob1, encoded_df], axis=1)
df_ob1=df_ob1.drop(['race'], axis=1)

##encoding for age_group column 
data_encoded_1 = encoder.fit_transform(df_ob1[['age_group']])
categories_1 = encoder.categories_[0]
encoded_df_1= pd.DataFrame(data_encoded_1, columns=categories_1)
df_ob1 = pd.concat([df_ob1, encoded_df_1], axis=1)
df_ob1=df_ob1.drop(['age_group'], axis=1)

##encoding for  employment_status 
data_encoded_2 = encoder.fit_transform(df_ob1[['employment_status']])
categories_2 = encoder.categories_[0]
encoded_df_2= pd.DataFrame(data_encoded_2, columns=categories_2)
df_ob1 = pd.concat([df_ob1, encoded_df_2], axis=1)
df_ob1=df_ob1.drop(['employment_status','Other or Multiple'], axis=1)
last_index=df_ob1.columns[-1]
df_ob1.drop([last_index], axis=1,inplace=True)
df_ob1
```

```python
##--a function to find number of null values in each feature 
def null_values(data):
    feature_name=[]
    for feature in data.columns:
        a=data[feature].isna().sum()
        feature_name.append(feature)
        feature_name.append(a)
    return feature_name
```

```python
df_fi=pd.concat([df_ob1,df_ob2],axis=1)

```

```python
##  Handling missing/nan values in all columns----
def replace_miss_val(data):
    for feature in data.columns:
        data[feature]=data[feature].fillna(data[feature].mode()[0])
    return data
```

```python
##-- replacing the missing values with mode value of each column 
X=replace_miss_val(df_fi)
X=X.drop(X.columns[-2:], axis=1)

X.drop(columns=['respondent_id'], inplace=True)
y.drop(columns=['respondent_id'], inplace=True)
```
